## Постановка задачи

<div class="text-justify">
&ensp;
&#8194;
Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали.
</div>
<div class="text-justify">
&ensp;
&#8194;
Необходимо построить модель, которая предскажет температуру стали.
</div>

## Описание этапа обработки

<div class="text-justify">
&ensp;
&#8194;
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.
</div>
<div class="text-justify">
&ensp;
&#8194;
Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).
</div>
<div class="text-justify">    
&ensp;
&#8194;
Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.
</div>
<div class="text-justify">    
&ensp;
&#8194;
Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).
</div>

Для лучшего понимания построения модели изучим процесс металлообработки. В данном случае используется установка **Ковш-печь**.

Википедия: [Агрегат «ковш-печь»](https://ru.wikipedia.org/wiki/%D0%90%D0%B3%D1%80%D0%B5%D0%B3%D0%B0%D1%82_%C2%AB%D0%BA%D0%BE%D0%B2%D1%88-%D0%BF%D0%B5%D1%87%D1%8C%C2%BB)

![](https://metallplace.ru/upload/metphoto/kovshpech.jpg)


**Факты:**

<div class="text-justify">
    
+ Наличие этих установок позволяет понизить температуру процесса, тем самым улучшая производительность печей для плавления стали и способствуют улучшению огнеупорных свойств кладки печи.
+ Подобные установки стабилизируют температуру металла, повышая, при этом, вероятность изготовления продукта с необходимым химическим составом.
+ Вместительность данных печей лежит в пределах 30 - 150 т.
+ Скорость прогревания металла чаще всего составляет около 5 °С/мин.
+ Оптимальная температура, которую необходимо поддерживать в ковше, равняется примерно 1600 °С.
+ Необходимо поддерживать восстановительную атмосферу, исключив, при этом попадание в нее кислорода. Для этого под свод поступает инертный газ.
+ Общее время, занимаемое процессом, составляет от 60 до 90 минут. От 10 до 30 минут, при этом, уходит на дуговой нагрев, происходящий циклически.

</div>

**Можно применять следующие меры для улучшения обработки стали и уменьшения энергозатрат:**

<div class="text-justify">
    
+ Рациональное применение высокой температуры отработанного шлака – для нагрева ковшей, шлакообразующей шихты и других элементов.
+ Применение порошкообразных материалов в процессе продувки аргоном сквозь другую фурму – способствует избавлению от серы и повышает скорость раскисления металла. Уменьшается длительность процесса – снижаются затраты энергии.
+ Усиленное смешивание шлака со сталью при помощи перемешивающего индуктора – также сокращает длительность обработки.
    
</div>

### 2. Функции автоматизации

<div class="text-justify">
&ensp;
&#8194;
Создадим функции автоматизации. Начнем с поиска гиперпараметров.
</div>

```python
# функция для поиска гиперпараметров
def best_model(model, params, features_train, target_train):
    grid = GridSearchCV(model, param_grid = params, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1, verbose=1)
    grid.fit(features_train, target_train)
    print(grid.best_score_)
    print(grid.best_params_)
```

<div class="text-justify">
&ensp;
&#8194;
Обучение и подсчет времени.
</div>

```python
# функция для обучения модели
def fit_model(model, features_train, target_train):
    start_time = time.time()
    model.fit(features_train, target_train)
    finish_time = time.time() - start_time
    return round(finish_time, 3)
```

<div class="text-justify">
&ensp;
&#8194;
Подсчет скоринга на кросс-валидации.
</div>

```python
# функция для подсчета скорринга
def cv_score(model, features_train, target_train):
    score = cross_val_score(model, features_train, target_train, cv=5, scoring='neg_mean_absolute_error')
    return score.mean().round(3)
```

<div class="text-justify">
&ensp;
&#8194;
Предикт и MAE.
</div>

```python
# функция для предсказания
def prediction(model, features, target):   
    mae = mean_absolute_error(target, model.predict(features))   
    return mae.round(3)
```

<div class="text-justify">
&ensp;
&#8194;
Представление результатов.
</div>

```python
# функция для отображения результатов
def results_train(row, model, time_learn, mae_train):
    results.loc[row, 'model'] = model
    results.loc[row, 'time_learn, s'] = time_learn
    results.loc[row, 'cross_val'] = mae_train
    display(results)
```
